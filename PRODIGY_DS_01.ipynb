```python
#data analysis libraries
import numpy as np
import pandas as pd

#visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#ignore warnings
import warnings
```

```python
population=pd.read_csv('/content/archive (5).zip')
```

```python
population.columns
```

```python
print(population['2022 Population'].describe())
```

```python
plt.hist(population['2022 Population'], bins=10)
plt.title('Histogram of World Population')
plt.xlabel('Population (in billions)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.scatter(population['Area (km²)'], population['Density (per km²)'])
plt.title('Distribution of World Area and Population Density')
plt.xlabel('Area (in square kilometers)')
plt.ylabel('Population Density (per square kilometer)')
plt.show()
```

```python
plt.hist(population['2022 Population']/1000000000, bins=15)
plt.title('Histogram of World Population')
plt.xlabel('Population (in billions)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.boxplot(population['2022 Population']/100000000)
plt.title('Box Plot of World Population')
plt.ylabel('Population (in 100 million)')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data: Gender distribution
genders = ['Male', 'Female', 'Non-Binary', 'Other']
counts = [300, 400, 50, 30]

# Create a bar chart
plt.bar(genders, counts, color=['blue', 'pink', 'purple', 'gray'])

# Add labels and title
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution in a Population')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random ages for demonstration purposes
np.random.seed(42)
ages = np.random.randint(18, 65, 1000)

# Create a histogram
plt.hist(ages, bins=20, color='blue', edgecolor='black')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Ages in a Population')

# Show the plot
plt.show()
```

```python
data_by_continent = population.groupby('Continent').sum()

plt.pie(data_by_continent['2022 Population'], labels=data_by_continent.index, autopct='%1.1f%%')
plt.title('World Population by Continent')
plt.show()
```
